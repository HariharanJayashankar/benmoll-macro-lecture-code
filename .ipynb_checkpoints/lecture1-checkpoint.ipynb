This follows [this lecture](https://benjaminmoll.com/wp-content/uploads/2019/07/Lecture1_2149.pdf).

Particularly, it implements the shooting algorithm in the end. I modified the algorithm given a bit so that we can update the values of $c^0$ slightly better.

The two *Finite Difference* equations (which are the approximations to the ODEs) we aim to solve are:

$$
c^{n+1} = \Delta t c^n \frac{1}{\sigma} (F'(k^n) - \rho - delta) + c^n \tag{FD-1}
$$

$$
k^{n+1} = \Delta t (F(k^n) - \delta k^n - c^n) + k_n \tag{FD-2}
$$

With $k^0$ given as $\frac{k*}{2}$

## The Solution Algorithm

Like the lecturre notes, I use the shooting algorithm. The algorithm in the notes doesn't have a proper rule for updatig $c^0$ - our guess of the initial consumption value. So I implement a slightly modified version of it.

Define the upper bound of $c^0$ as $c^{upper} := F(k^0)$ and the lower bound as $c^{lower} := 0$.

Now:
1. Let $c^0 := \frac{c^{upper} + c^{lower}}{2}$
2. Implement the finite difference equations to get $(c^n, k^n)$ for $n \in (1, 2,. ... N)$
3. Check whether $k^n$ is equal to $k*$
4. If $k_n > k*$, $c^0$ is too low. Therefore we can update our new $c^{lower} := c^0$. If $k_n < k*$, $c^0$ is too high therefore we can update $c^{upper} := c^0$

```julia
using LinearAlgebra, Parameters, Plots
```

```julia
# Storing parameters
opgrowth = @with_kw (α = 0.3, 
    A = 1, σ = 2, ρ = 0.05, 
    δ = 0.05, N = 700,
    Δt = 0.1)

function u(c, σ = 2.0)
    # CRRA utility
   
    if σ == zero(σ)
        u = log(c)
    else
        u = (c^(1 - σ))/(1 - σ)
    end
    
    return u
end

# Some nice formulas based on parameters
F(k, α = 0.3, A = 1.0) = A*k^α
Fdash(k, α = 0.3, A = 1.0) = A * α * k ^ (α - 1)
kstar(α = 0.3, A = 1.0, ρ = 0.05, δ = 0.05) = (α * A/(ρ + δ))^(1/(1-α))
cstar(α = 0.3, A = 1.0, ρ = 0.05, δ = 0.05) = F(kstar(α, A, ρ, δ), α, A) - δ * kstar(α, A, ρ, δ)

# Finite Difference equations
c_fd(kn, cn, α = 0.3, A = 0.1, σ = 2.0,
    Δt = 0.1, ρ = 0.05, δ = 0.05) = Δt*cn * (1/σ) * (Fdash(kn, α, A) - ρ - δ) + cn
k_fd(kn, cn,  α = 0.3, A = 0.1,
    Δt = 0.1, δ = 0.05) = Δt * (F(kn, α, A) - δ * kn - cn) + kn
```

```julia
# Functions to implement shooting
function create_series!(cseries, kseries, cinit, kinit, params)
    # creates time series of c and k 
    # it takes an existing empty time series to
    # modify it in place
    
    @unpack α, A, σ, ρ, δ, N, Δt = params
    c0 = cinit
    k0 = kinit
    
    for n = 1:N
        cseries[n] = c0
        kseries[n] = k0

        c0 = c_fd(k0, c0, α, A, σ, Δt, ρ, δ)
        k0 = k_fd(k0, c0, α, A, Δt, δ)
        
        # replace k0, c0 with a very small number if it is not greater than 0
        k0 = k0 > 0 ? k0 : 1e-5
        c0 = c0 > 0 ? c0 : 1e-5
                
        k0 = isnan(k0) ? 0 : k0
        c0 = isnan(c0) ? 0 : c0
    end
    
end
```

```julia
function shoot(params, cinit = 1e-6, kinit = kstar()/2.0,
                maxiter = 1000, tol = 1e-6)
    
    @unpack α, A, σ, ρ, δ, N, Δt = params
    
    kfinal = kstar(α, A, ρ, δ)
    cfinal = cstar(α, A, ρ, δ)
    
    # initiate empty matrices to store values in
    cseries = zeros(N)
    kseries = zeros(N)
     
    iter = 0
    error = kinit - kfinal
    
    # define upper and lower bounds
    clb = 0.0
    cub = F(kinit)
    
    while abs(error) > tol && iter < maxiter
        
        # cinit is the mid point of our bounds
        cinit = (clb + cub)/2.0

        create_series!(cseries, kseries, cinit, kinit, params)
        
        error = kseries[end] - kfinal
        
        # Update upper and lower bounds of c
        if error > 0
            clb = cinit
        else
            cub = cinit
        end
        
        iter += 1
        
    end
    
    # Output Message
    if iter == maxiter
        println("==========================", "\n",
                "Convergence failed. Max iterations of $maxiter reached", "\n",
                "Error: $error", "\n",
                "==========================")
    else
        println("==========================", "\n",
                "Convergence Successful!", "\n",
                "Error: $error", "\n",
                "Iterations: $iter", "\n",
                "==========================")
    end
    
    return (cseries, kseries, error, iter)
    
    
end
```

```julia
og = opgrowth()
cseries, kseries, error, iter = shoot(og)
```

```julia
plot(cseries, label = "", legend=:bottomright)
Plots.abline!(0, cstar(), line=:dash, label = "c*", color = "black")
plot!(title = "Consumption")
```

```julia
plot(kseries, label = "", legend=:bottomright)
Plots.abline!(0, kstar(), line=:dash, label = "k*", color = "black")
plot!(title = "Capital", xlabel = "Time")
```
